```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#sns.set()
```

```python
sns.set()
df_raw_data = pd.read_excel("MovieSurvey (Responses).xlsx")
```

```python
#Good practice, back data up.
df_survey_result = df_raw_data.copy()
df_survey_result
```

```python
#Rename columns
new_column_names=['Timestamp','Age','Residence','Genre','Frequency','Preferred_Platform','Schedule','Companions','Selling Point','Likes','Dislike','References']
df_survey_result.columns = new_column_names
```

```python
#You need to santise columns that are multiselect.  
#Columns that are not multi-select, use them as-is
non_multi_select_columns = ['Age','Residence','Preferred_Platform','Frequency','References']

df_survey_partial_data = df_survey_result[(non_multi_select_columns)]
```

```python
#create a function to pre-process data
def process_data(input_df,source_df,column_array,col_to_assess):
    for row in range(input_df.shape[0]):
        for x in range(len(column_array)):
            input_df.iloc[row,input_df.columns.get_loc(column_array[x])] = (1 if source_df.iloc[row,col_to_assess].find(column_array[x])!=-1 else 0)
           # print(column_array[x] + "compare with " + source_df.iloc[row,col_to_assess] + "result is :" +
           #       str(source_df.iloc[row,col_to_assess].find(column_array[x])))
           # print(source_df.iloc[row,col_to_assess])
           # print(x)
    return input_df

#split muti select column values into columns
#splitting Genre
#known Genres : Action, Adventure, Comedy,Horror,Scifi,Fantasy,Mystery,Thriller,Crime,
#Romance,Family,Documentary,Musical,Bioagraphy,Historical
genre_columns = ['Action','Adventure','Comedy','Horror','Sci-Fi','Fantasy',
'Mystery','Thriller','Crime','Romance','Family',
'Documentary','Musical','Biography','Historical']

df_blank_genre_columns = pd.DataFrame(columns=genre_columns)
#concatenate main data with new columns
df_temp=pd.concat([df_survey_partial_data,df_blank_genre_columns],axis=1)
#pre-process data
df_temp=process_data(df_temp,df_survey_result,genre_columns,3)
df_survey_partial_data_w_genre = df_temp
df_survey_partial_data_w_genre
```

```python
#split muti select column values into columns
#
#Schedules : After Work, Weekends, Special Occasions, Others
schedule_columns = ['Afterwork', 'Weekends','Special Occasions','Others']
df_blank_schedule_column = pd.DataFrame(columns=schedule_columns)
df_temp=pd.concat([df_survey_partial_data_w_genre,df_blank_schedule_column],axis=1)
df_temp
df_temp=process_data(df_temp,df_survey_result,schedule_columns,6)
df_survey_partial_data_w_genre_schedule = df_temp
df_survey_partial_data_w_genre_schedule
```

```python
#Split companions
companions = ['Family', 'Friends','Partner/Spouse','Alone','Others']
df_companion_columns = pd.DataFrame(columns=companions)
df_temp=pd.concat([df_survey_partial_data_w_genre_schedule,df_companion_columns],axis=1)
df_temp
df_temp=process_data(df_temp,df_survey_result,companions,7)
df_survey_partial_data_w_genre_schedule_comp = df_temp
df_survey_partial_data_w_genre_schedule_comp
```

```python
#create decision columns
decision_factor = ['Trailer', 'Reviews','Suggestions','Plot','Actors_Actress','Others']
decision_factor_column = pd.DataFrame(columns=decision_factor)
df_temp=pd.concat([df_survey_partial_data_w_genre_schedule_comp,decision_factor_column],axis=1)


df_temp=process_data(df_temp,df_survey_result,decision_factor,8)
df_survey_partial_data_w_genre_schedule_comp_factor = df_temp
df_survey_partial_data_w_genre_schedule_comp_factor
```

```python
like_columns = ['Plot/Story', 'Soundtrack','Visual Effects','Acting','Actors/Actresses']
df_like_columns = pd.DataFrame(columns=like_columns)
df_temp=pd.concat([df_survey_partial_data_w_genre_schedule_comp_factor,df_like_columns],axis=1)
#df_temp
df_survey_partial_data_w_genre_schedule_comp_factor_like=process_data(df_temp,df_survey_result,like_columns,9)
df_survey_partial_data_w_genre_schedule_comp_factor_like
```

```python
dislike_columns = ['Plot/Story', 'Soundtrack','Visual Effects','Acting','Actors/Actresses']
df_dislike_columns = pd.DataFrame(columns=dislike_columns)
df_temp=pd.concat([df_survey_partial_data_w_genre_schedule_comp_factor_like,df_dislike_columns],axis=1)
#df_temp
df_survey_partial_data_w_genre_schedule_comp_factor_like_dislke=process_data(df_temp,df_survey_result,dislike_columns,10)
df_survey_partial_data_w_genre_schedule_comp_factor_like_dislke
```

```python
df_sanitized_data=df_survey_partial_data_w_genre_schedule_comp_factor_like_dislke
```

```python
new_column_names = ['Age', 'Residence', 'Preferred_Platform', 'Frequency', 'References',
       'Action', 'Adventure', 'Comedy', 'Horror', 'Sci-Fi', 'Fantasy',
       'Mystery', 'Thriller', 'Crime', 'Romance', 'Family', 'Documentary',
       'Musical', 'Biography', 'Historical', 'Afterwork', 'Weekends',
       'Special Occasions', 'Others', 'C_Family', 'Friends', 'Partner/Spouse',
       'Alone', 'Others Companions', 'Trailer', 'Reviews', 'Suggestions', 'Plot',
       'Actors/Actress', 'Others', 'L_Story', 'L_Soundtrack',
       'L_Visual Effects', 'L_Acting', 'L_Actors/Actresses', 'D_Plot/Story',
       'D_Soundtrack', 'D_Visual Effects', 'D_Acting', 'D_Actors/Actresses']
df_sanitized_data.columns = new_column_names
```

```python
#VISUALIZE
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()

age_group_series = df_sanitized_data['Age'].value_counts()

pie_labels = df_sanitized_data['Age'].unique()
plt.figure(figsize=(10,7))
plt.pie(age_group_series,labels= df_sanitized_data['Age'].unique(),
       autopct='%.2f%%', 
        textprops={ 'fontweight' : 'bold','rotation' : 35,'rotation':35})
plt.legend(pie_labels,loc="upper left")
plt.title('Respondents According to Age Groups')
plt.savefig('LJ_pie.png')
plt.show()
```

```python
all_columns=['Age','Action','Adventure','Comedy','Horror','Sci-Fi','Fantasy',
'Mystery','Thriller','Crime','Romance','Family',
'Documentary','Musical','Biography','Historical']
categories  = ['Action','Adventure','Comedy','Horror','Sci-Fi','Fantasy',
'Mystery','Thriller','Crime','Romance','Family',
'Documentary','Musical','Biography','Historical']

df_bar_chart = df_sanitized_data[all_columns]
#summarise data
df_bar_chart_sum=df_bar_chart.groupby('Age')[categories].sum().reset_index()
df_bar_chart_sum
```

```python
df_bar_chart_sum.set_index('Age',inplace=True)
df_bar_T = df_bar_chart_sum.transpose()
df_bar_T= df_bar_T.rename_axis('Genre')
df_bar_T
```

```python
ax = df_bar_T.plot(kind='barh', stacked=False, figsize=(12, 8),width=2,edgecolor='black')
ax.set_xlabel('Count')
ax.set_ylabel('Genres')
#ax.set_xticks(range(df_bar_T.max().max()))
ax.set_title('Movie Genres Preferences by Age Group')
plt.savefig('LJ_BARH_preferency_by_age_group.png')
plt.show()
```

```python
#Preferred platform by Age Group
df_frequency = df_sanitized_data.copy()
df_frequency = df_frequency[(['Frequency','Age'])]
df_frequency=df_frequency.groupby(['Frequency','Age']).size().reset_index()
#df_frequency_ri = df_frequency.set_index('Frequency').reset
#df_frequency_ri
df_frequency.columns = ['Frequency','Age','Count']
df_frequency
```

```python
df_age_frequency_pivot = df_frequency.pivot(index='Frequency',columns='Age',values='Count').fillna(0)
df_age_frequency_pivot
```

```python
ax = df_age_frequency_pivot.plot(kind='barh',stacked=False,align='center',edgecolor='black', figsize=(20, 8))
ax.set_xticks(range(int(df_age_frequency_pivot.max().max())+1))
ax.set_xlabel('Count per Age Group')
ax.set_ylabel('Frequency')
ax.set_title('Viewing Habbits by Age Group')

#ax.set_xticks(range(len()))
#plt.title("Cars Listings by Brand", fontsize = 16, fontweight = "bold")
#plt.ylabel("Number of Listings", fontsize=13)
plt.savefig("LJ_barh_viewing_habbits_by_age.png")
plt.show()



#df_test=df_age_frequency_pivot.reset_index()
#df_test
```
