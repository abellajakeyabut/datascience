```python
base_url="http://api.exchangeratesapi.io/{}?symbols={}&access_key={}"
access_key = "cd87ff01651a3172b3ec35839ebfa7ce"
date = input("Please key-in date in format yyyy-mm-dd or latest")
base = input("Currency to convert")
curr = input("Destination currency")
quantity = float(input("How much {} do you want to convert: ".format(base)))
url = base_url.format(date,curr,access_key)
convert = lambda x,y: x*y

response = requests.get(url)
if(response.ok is False):
    print("Error {}".format(response.status_code))
    print("response {}".format(response.json()['error']))
else:
    data = response.json()
    rate = data['rates'][curr]
    result = convert(quantity, rate)

print("{} {} is equal to {} {}".format(quantity,base,result,curr))
response.json()

    
```
