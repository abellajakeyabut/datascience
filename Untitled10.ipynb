```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#sns.set()
```

```python
df_used_cars = pd.read_csv("bar_chart_data.csv")
```

```python
df_used_cars
```

```python
plt.figure(figsize=(9,6))
plt.bar(x = df_used_cars['Brand'],height = df_used_cars['Cars Listings'],
       color = "midnightblue")
plt.xticks(rotation=45)
plt.title("Cars Listings by Brand", fontsize = 16, fontweight = "bold")
plt.ylabel("Number of Listings", fontsize=13)
plt.savefig("bar.png")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_fuel_engine_type = pd.read_csv("pie_chart_data.csv")
```

```python
df_fuel_engine_type
```

```python
plt.figure(figsize=(10,8))
plt.pie(df_fuel_engine_type['Number of Cars'],labels=df_fuel_engine_type['Engine Fuel Type'].values,autopct='%.2f%%'
       ,textprops={'size' : 'x-large', 'fontweight' : 'bold','rotation' : 35,'color' : 'w'})
plt.title('Cars by Engine Fuel Type',fontsize=18,fontweight = 'bold')
plt.legend()
plt.savefig("pie.png")
plt.show()
```
