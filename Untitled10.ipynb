```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#sns.set()
```

```python
df_used_cars = pd.read_csv("bar_chart_data.csv")
```

```python
df_used_cars
```

```python
plt.figure(figsize=(9,6))
plt.bar(x = df_used_cars['Brand'],height = df_used_cars['Cars Listings'],
       color = "midnightblue")
plt.xticks(rotation=45)
plt.title("Cars Listings by Brand", fontsize = 16, fontweight = "bold")
plt.ylabel("Number of Listings", fontsize=13)
plt.savefig("bar.png")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_fuel_engine_type = pd.read_csv("pie_chart_data.csv")
```

```python
df_fuel_engine_type
```

```python
plt.figure(figsize=(10,8))
plt.pie(df_fuel_engine_type['Number of Cars'],labels=df_fuel_engine_type['Engine Fuel Type'].values,autopct='%.2f%%'
       ,textprops={'size' : 'x-large', 'fontweight' : 'bold','rotation' : 35,'color' : 'w'})
plt.title('Cars by Engine Fuel Type',fontsize=18,fontweight = 'bold')
plt.legend()
plt.savefig("pie.png")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
sns.set_style("white")
```

```python
df_stacked_area_chart = pd.read_csv('stacked_area_chart_data.csv')
```

```python
df_stacked_area_chart
```

```python
colors = ["#011638","#7e2987", "#ef2026"]
labels = ["Desel","Petrol","Gas"]
plt.figure(figsize=(12,6))
plt.stackplot(df_stacked_area_chart['Year'],
             df_stacked_area_chart['Diesel'],
             df_stacked_area_chart['Petrol'],
             df_stacked_area_chart['Gas'],
             colors=colors,
             edgecolor ='none')
plt.legend(labels = labels,loc ="upper left")
plt.ylabel("Number of Cars",fontsize = 13)
sns.despine()
plt.xticks(df_stacked_area_chart['Year'],rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

```python
df_spx_ftse_00_10 = pd.read_csv("line_chart_data.csv")
```

```python
df_spx_ftse_00_10
```

```python
df_spx_ftse_00_10["new_date"] = pd.to_datetime(df_spx_ftse_00_10["Date"],format="%m/%d/%Y")
```

```python
df_spx_ftse_00_10
```

```python
labels = ["S&P 500","FTSE 100"]
plt.figure(figsize=(20,8))
plt.plot(df_spx_ftse_00_10["new_date"],df_spx_ftse_00_10["GSPC500"])
plt.plot(df_spx_ftse_00_10["new_date"],df_spx_ftse_00_10["FTSE100"])
plt.title("S&P vs FTSE Retunrs (2000-2010)",fontsize=14,fontweight="bold")
plt.ylabel("Returns")
plt.xlabel("Date")
plt.legend(labels=labels,fontsize="large")
```

```python
df_spx_ftse_H2_08 = df_spx_ftse_00_10[(df_spx_ftse_00_10.new_date >='2008-07-01') &
                                     (df_spx_ftse_00_10.new_date <= '2008-12-31')]
```

```python
df_spx_ftse_H2_08
```

```python
labels = ["S&P 500","FTSE 100"]
plt.figure(figsize=(20,8))
plt.plot(df_spx_ftse_H2_08["new_date"],df_spx_ftse_H2_08["GSPC500"])
plt.plot(df_spx_ftse_H2_08["new_date"],df_spx_ftse_H2_08["FTSE100"])
plt.title("S&P vs FTSE Retunrs (H2 2008)",fontsize=14,fontweight="bold")
plt.ylabel("Returns")
plt.xlabel("Date")
plt.legend(labels=labels,fontsize="large")
plt.savefig("2008_Quarter_2_line.png")
plt.show()
```
